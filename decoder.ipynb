```python
#DECODER
import mp3
import os
from wave import Wave_write


carpeta = r'MEMD_audio'
archivos_en_carpeta = os.listdir(carpeta)
for archivo in archivos_en_carpeta:
    with open(os.path.join(carpeta, archivo), 'rb') as read_file, open(archivo[:-4] + '.wav', 'wb') as write_file:
        decoder = mp3.Decoder(read_file)
        sample_rate = decoder.get_sample_rate()
        nchannels = decoder.get_channels()

        wav_file = Wave_write(write_file)
        wav_file.setnchannels(nchannels)
        wav_file.setsampwidth(2)
        wav_file.setframerate(sample_rate)

        while True:
            pcm_data = decoder.read(4000)

            if not pcm_data:
                break
            else:
                wav_file.writeframes(pcm_data)
```
